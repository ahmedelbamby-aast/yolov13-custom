Runs S/L/X benchmark comparisons with fixed largest batches (S=32, L=16, X=12)
for both Turing Flash backend and fallback backend using normal Python/Ultralytics API flow.


```python
from pathlib import Path
import os, sys, shutil, subprocess

REPO_URL = 'https://github.com/ahmedelbamby-aast/yolov13-custom.git'
WORKING_REPO = Path('/kaggle/working/yolov13-custom')
DEV_REPO = Path('/kaggle/work_here/yolov13')

if not WORKING_REPO.exists():
    subprocess.check_call(['git', 'clone', REPO_URL, str(WORKING_REPO)])
if not DEV_REPO.exists():
    DEV_REPO.parent.mkdir(parents=True, exist_ok=True)
    shutil.copytree(WORKING_REPO, DEV_REPO, dirs_exist_ok=True)

os.chdir(DEV_REPO)
if str(DEV_REPO) not in sys.path:
    sys.path.insert(0, str(DEV_REPO))
os.environ['LD_LIBRARY_PATH'] = '/usr/local/nvidia/lib64:/usr/local/cuda/lib64:' + os.environ.get('LD_LIBRARY_PATH','')
os.environ.setdefault('Y13_USE_TURING_FLASH', '1')
os.environ.setdefault('Y13_DISABLE_FLASH', '0')
print('repo:', DEV_REPO)
```

```python
# Run two fixed-batch benchmark passes: Turing flash and fallback
import importlib.util
from pathlib import Path

p = Path(kaggle/scripts/170_benchmark_slx_fixed_batch.py)
spec = importlib.util.spec_from_file_location(bench_slx_fixed, p)
bench = importlib.util.module_from_spec(spec)
spec.loader.exec_module(bench)

summary_turing = bench.run(
    data_yaml=coco128.yaml,
    epochs=30,
    workers=4,
    use_turing_flash=True,
    force_disable_flash=False,
    run_name=y13_bench_slx_30e_turing_fixed_v2,
)
summary_fallback = bench.run(
    data_yaml=coco128.yaml,
    epochs=30,
    workers=4,
    use_turing_flash=False,
    force_disable_flash=True,
    run_name=y13_bench_slx_30e_turing_fixed,
)
print(turing backend:, summary_turing[flash_backend])
print(fallback backend:, summary_fallback[flash_backend])
```
