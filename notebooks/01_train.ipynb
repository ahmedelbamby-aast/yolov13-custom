# 01 - Train (Kaggle UI, no SSH)
This notebook prepares the repo and runs DDP training from Kaggle UI.

```python
from pathlib import Path
import os
import shutil
import subprocess

REPO_URL = 'https://github.com/ahmedelbamby-aast/yolov13-custom.git'
WORKING_REPO = Path('/kaggle/working/yolov13-custom')
DEV_REPO = Path('/kaggle/work_here/yolov13-custom')

if not WORKING_REPO.exists():
    subprocess.check_call(['git', 'clone', REPO_URL, str(WORKING_REPO)])

DEV_REPO.parent.mkdir(parents=True, exist_ok=True)
if not DEV_REPO.exists():
    shutil.copytree(WORKING_REPO, DEV_REPO, dirs_exist_ok=True)

os.chdir(DEV_REPO)
os.environ['LD_LIBRARY_PATH'] = '/usr/local/nvidia/lib64:/usr/local/cuda/lib64:' + os.environ.get('LD_LIBRARY_PATH', '')
print('Repository ready at', DEV_REPO)
```

```python
import subprocess
subprocess.check_call(['bash', 'kaggle/scripts/10_setup_uv.sh'])
subprocess.check_call(['bash', 'kaggle/scripts/20_install_deps.sh'])
subprocess.check_call(['bash', 'kaggle/scripts/30_gpu_check.sh'])
print('Environment setup done')
```
